# Titanic without using pipelines

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
# drop ['PassengerId','Name','Ticket','Cabin'] columns as they are not for use 
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
df.head()
```

```python
# Step 1 -> train/test/split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['Survived']),
                                                df['Survived'],
                                                test_size=0.2,
                                                random_state=42)
```

```python
X_train.head(2)
```

```python
y_train.head(4)
```

```python
# have missing values present
df.isnull().sum()
```

```python
# applying imputation to remove missing values

si_age = SimpleImputer() # missing value in age replace with mean as Simple imputator replace missing values by default mean
si_embarked = SimpleImputer(strategy='most_frequent') # missing value in Embarked replace with most frequent values

X_train_age = si_age.fit_transform(X_train[['Age']])
X_train_embarked = si_embarked.fit_transform(X_train[['Embarked']])

X_test_age = si_age.fit_transform(X_test[['Age']])
X_test_embarked = si_embarked.fit_transform(X_test[['Embarked']])
```

```python
X_train_age
```

```python
X_train_embarked
```

```python
# convert sex and exbarked (categorical variables) to numerical using one hot encoding

ohe_sex = OneHotEncoder(sparse=False,handle_unknown='ignore')
ohe_embarked = OneHotEncoder(sparse=False,handle_unknown='ignore')

X_train_sex = ohe_sex.fit_transform(X_train[['Sex']])
X_train_embarked = ohe_embarked.fit_transform(X_train_embarked)

X_test_sex = ohe_sex.fit_transform(X_test[['Sex']])
X_test_embarked = ohe_embarked.fit_transform(X_test_embarked)
```

```python
X_train_sex
```

```python
X_train_embarked
```

```python
X_train_rem = X_train.drop(columns=['Sex','Age','Embarked'])
```

```python
X_test_rem = X_test.drop(columns=['Sex','Age','Embarked'])
```

```python
X_train_transformed = np.concatenate((X_train_rem,X_train_age,X_train_sex,X_train_embarked),axis=1)
X_test_transformed = np.concatenate((X_test_rem,X_test_age,X_test_sex,X_test_embarked),axis=1)
```

```python
X_train_transformed
```

```python
X_train_transformed.shape
```

```python
X_test_transformed
```

```python
X_test_transformed.shape
```

```python
clf=DecisionTreeClassifier()
clf.fit(X_train_transformed,y_train)
```

```python
y_pred = clf.predict(X_test_transformed)
y_pred
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test,y_pred)
```

# Titanic Using Pipelines

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline,make_pipeline
from sklearn.feature_selection import SelectKBest,chi2
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

# Let's Plan

```python
# drop ['PassengerId','Name','Ticket','Cabin'] columns as they are not for use 
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
# Step 1 -> train/test/split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['Survived']),
                                                df['Survived'],
                                                test_size=0.2,
                                                random_state=42)
```

```python
X_train.head()
```

```python
y_train.sample(5)
```

```python
# imputation transform
trf1 = ColumnTransformer([
    ('impute_age',SimpleImputer(),[2]),
    ('impute_embarked',SimpleImputer(strategy='most_frequent'),[6])
],remainder='passthrough')
```

```python
# One hot encoding
trf2 = ColumnTransformer([
    ('ohe_sex_embarked',OneHotEncoder(sparse=False,handle_unknown='ignore'),[1,6])
],remainder='passthrough')
```

```python
# Scaling
trf3 = ColumnTransformer([
    ('scale',MinMaxScaler(),slice(0,10))
])
```

```python
# Feature Selection
trf4 = SelectKBest(score_func=chi2,k=8)
```

```python
# train the model
trf5 = DecisionTreeClassifier()
```

# Create Pipelines

```python
pipe = Pipeline([
    ('trf1',trf1),
    ('trf2',trf2),
    ('trf3',trf3),
    ('trf4',trf4),
    ('trf5',trf5)
])
```

# Pipeline vs make_pipeline

### Pipeline requires naming of steps, make_pipeline does not.
### (Sample applies to Column Transformer vs make_column_transformer)

```python
# alternate you can also use make_pipelinf func b/c it is easy to implement
pipe = make_pipeline(trf1,trf2,trf3,trf4,trf5)
```

```python
# train
pipe.fit(X_train,y_train)
```

# Explore the Pipeline

```python
# code here
pipe.named_steps
```

```python
pipe.named_steps['trf1'].transformers_
```

```python
pipe.named_steps['trf1'].transformers_[0][1].statistics_
```

```python
pipe.named_steps['trf1'].transformers_[1][1].statistics_
```

```python
# Display pipeline

from sklearn import set_config
set_config(display='diagram')
```

```python
# Predict 
y_pred = pipe.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

# Cross validation using Pipeline

```python
# cross validation using cross_val_score
from sklearn.model_selection import cross_val_score
cross_val_score(pipe,X_train,y_train,cv=5,scoring='accuracy').mean()
```

# GridSearch using Pipeline

```python
# gridsearchcv (hyperparameter tuning)
params = {
    'trf5__max_depth':[1,2,3,4,5,None]
}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(pipe,params,cv=5,scoring='accuracy')
grid.fit(X_train,y_train)
```

```python
grid.best_score_
```

```python
grid.best_params_
```
